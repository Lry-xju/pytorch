## L4 BP

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
x_data = [1,2,3]
y_data = [2,4,6]
w  = torch.Tensor([1.0])
w.requires_grad = True;# w
def forward(x):
    # wtensor
    return x*w
# 
def loss(x,y):    
    y_pred = forward(x);
    return (y_pred - y)**2
print('predict (before training )',4,forward(4).item())
epoch_list = []
cost_list =[]

for epoch in range(100):
    for x_val ,y_val in zip(x_data,y_data):
        l = loss(x_val,y_val) #forward
        l.backward() # lbp
        print('\t grad:',x_val,y_val,w.grad.item()) #w.grad.item() 
        w.data = w.data -0.01 * w.grad.data # ww.grad
        #w.grad.data lossww.data 
        #data &grad tensor
        print(' w.grad.data.type = ', type(w.grad.data))
        w.grad.data.zero_()# 
    print('epoch :',epoch,l.item())
    epoch_list.append(epoch)
    cost_list.append(l.item()) # l
print('predict (after training )',4,forward(4))
# 
plt.figure(figsize=(8, 4))  # 
plt.plot(epoch_list, cost_list, marker='o', linestyle='-', color='b', label='cost')

# 
plt.title('training')
plt.xlabel('epoch')
plt.ylabel('cost')
plt.legend()

# 
plt.grid(True)  # 
plt.show()
```

## L5 pytorch 

```python
class LinearModel(torch.nn.Module):
        def __init__(self):#
            super(LinearModel, self).__init__()
            self.linear = torch.nn.Linear(1, 1,bias = True)
            # self.linear = torch.nn.Linear(3, 3)
        # 
        def forward(self, x): 
            y_pred = self.linear(x)
            # y_pred = torch.sigmoid(self.linear(x))
            return y_pred
model = LinearModel()
criterion = torch.nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=1e-3) # Adam

x_data = torch.tensor([[1.0], [2.0], [3.0]])
y_data = torch.tensor([[2.0], [4.0], [6.0]])
epoch_list = []
cost_list =[]
for epoch in range(1000):
    y_pred = model(x_data)
    loss = criterion(y_pred, y_data)
    if epoch %200 ==0:
        print(epoch, loss.item())
    optimizer.zero_grad()
    loss.backward()#   
    optimizer.step() #    
    epoch_list.append(epoch)
    cost_list.append(loss.item()) # l
print ('w = ',model.linear.weight.item())
print ('b = ',model.linear.bias.item())

x =  torch.tensor([4.0])
x_data = torch.tensor([[1.0], [2.0], [3.0]])
print("",model(x_data))
print('predict (after training )',4,model(x))
# 
plt.figure(figsize=(8, 4))  # 
plt.plot(epoch_list, cost_list, marker='o', linestyle='-', color='b')

# 
plt.title('training')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()

# 
plt.grid(True)  # 
plt.show()
```

```python
import torch
# import torch.nn.functional as nnf
import numpy as np
import matplotlib.pyplot as plt

plt.grid()

x_data = torch.tensor([[1.0], [2.0], [3.0]])
y_data = torch.tensor([[0.0], [0.0], [1.0]])
# x_data = torch.tensor([1.0, 2.0, 3.0])
# y_data = torch.tensor([0.0, 0.0, 1.0])


class LogisticRegressionModel(torch.nn.Module):
    def __init__(self):
        super(LogisticRegressionModel, self).__init__()
        self.linear = torch.nn.Linear(1, 1)
        # self.linear = torch.nn.Linear(3, 3)

    def forward(self, x):
        y_pred = torch.sigmoid(self.linear(x))
        return y_pred


model = LogisticRegressionModel()

criterion = torch.nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=1e-3)

for epoch in range(1000):
    y_pred = model(x_data)
    loss = criterion(y_pred, y_data)
    if epoch %200 ==0:
        print(epoch, loss.item())

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

x = np.linspace(0, 10, 200)
x_t = torch.tensor(x).view(200, 1)
y_t = model(x_t.float())
y = y_t.data.numpy()

plt.plot(x, y)
plt.plot([0, 10], [0.5, 0.5], c='r')
plt.xlabel('Hours')
plt.ylabel('Probability of pass')
plt.savefig('FlaseGraph')
```
