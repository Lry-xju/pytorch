## L2 线性模型

```python
import numpy as np
import matplotlib.pyplot as plt
x_data = [1,2,3]
y_data = [2,4,6]
def forward(x):
    return x*w
def loss(x,y):
    y_pred = forward(x);
    return (y_pred - y)*(y_pred - y)
w_list = []
mse_list = []
for w in np.arange(0,4.1,0.1):
    l_sum =0
    print("w= ",w)
    for x_val,y_val in zip(x_data,y_data):
        y_pred_val = forward(x_val)
        loss_val = loss(x_val,y_val)
        l_sum += loss_val
        print ('\t',x_val,y_val,y_pred_val,loss_val)
    print("mse = ",l_sum/3)
    w_list.append(w)
    mse_list.append(l_sum/3.0)
```

## L3 GD 




```python
import numpy as np
import matplotlib.pyplot as plt
x_data = [1,2,3]
y_data = [2,4,6]
w = 1.0 
def forward(x):
    return x*w
#MSE ()
def cost(xs,ys):
    cost = 0
    for x,y in zip(xs,ys):
        y_pred = forward(x);
        cost +=(y_pred - y)**2
    return cost/len(xs)
#  cost对w求导
def gradient(xs,ys):
    grad= 0
    for x,y in zip(xs,ys):
        grad += 2 *x*(x*w-y)
        return grad /len(xs)
print('predict (before training )',4,forward(4))
epoch_list = []
cost_list =[]

for epoch in range(1000):
    cost_val = cost(x_data,y_data)
    grad_val = gradient(x_data,y_data)
    w = w -0.04 * grad_val
    print('epoch :',epoch,'w= ',w,'loss = ',cost_val)
    epoch_list.append(epoch)
    cost_list.append(cost_val)
print('predict (after training )',4,forward(4))
# 创建图形
plt.figure(figsize=(8, 4))  # 设置图形大小
plt.plot(epoch_list, cost_list, marker='o', linestyle='-', color='b', label='cost')

# 添加标题和标签
plt.title('training')
plt.xlabel('epoch')
plt.ylabel('cost')
plt.legend()

# 显示图形
plt.grid(True)  # 添加网格
plt.show()
```

## SGD

```python
import numpy as np
import matplotlib.pyplot as plt
x_data = [1,2,3]
y_data = [2,4,6]
w = 1.0 
def forward(x):
    return x*w
#MSE ()
def cost(x,y):    
    y_pred = forward(x);
    return (y_pred - y)**2

#  cost对w求导
def gradient(x,y): 
    return 2 *x*(x*w-y)

print('predict (before training )',4,forward(4))
epoch_list = []
cost_list =[]

for epoch in range(100):
    for x_val ,y_val in zip(x_data,y_data):
        cost_val = cost(x_val,y_val)
        grad_val = gradient(x_val,y_val)
        w = w -0.01 * grad_val
    print('epoch :',epoch,'w= ',w,'loss = ',cost_val)
    epoch_list.append(epoch)
    cost_list.append(cost_val)
print('predict (after training )',4,forward(4))
# 创建图形
plt.figure(figsize=(8, 4))  # 设置图形大小
plt.plot(epoch_list, cost_list, marker='o', linestyle='-', color='b', label='cost')

# 添加标题和标签
plt.title('training')
plt.xlabel('epoch')
plt.ylabel('cost')
plt.legend()

# 显示图形
plt.grid(True)  # 添加网格
plt.show()
```
