```python
import torch
from torchvision import transforms
from torchvision import datasets
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import torch.optim as optim
import torch.nn.functional as F
import numpy as np
```

```python
def show_images(images, labels , pred, nrows=4, ncols=4):
    """
    显示一个批次的图片和标签
    Args:
        images: torch.Tensor, 形状 [batch, 1, 28, 28]
        labels: torch.Tensor, 形状 [batch]
        nrows: 显示的行数
        ncols: 显示的列数
    """
    plt.figure(figsize=(12, 6))
    for i in range(nrows * ncols):
        plt.subplot(nrows, ncols, i+1)
        img = images[i].squeeze().cpu().numpy()  # 去掉通道维度 -> [28, 28]
        plt.imshow(img, cmap='gray')
        if labels[i].item()==pred[i].item():
            plt.title(f"✔ ,Label: {labels[i].item()}")
        else:
            plt.title(f" X ,Label: {labels[i].item()}, Pred: {pred[i].item()}")
        
        plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
import torchvision
batch_size = 64
transform  = transforms.Compose([
    transforms.ToTensor(), #convert pil image to tensor
    transforms.Normalize((0.1307,),( 0.3081,)) # mean & std,数字是对mnist计算出的经验值
])
train_dataset = datasets.MNIST(root='../dataset/mnist',train = True,transform=transform,download = True)
test_dataset = datasets.MNIST(root='../dataset/mnist',train = False,transform=transform,download = True)
train_loader = DataLoader(dataset= train_dataset,batch_size= batch_size ,shuffle= True)
test_loader = DataLoader(dataset= test_dataset,batch_size= batch_size ,shuffle= False)
```

```python
for batch_idx , data in enumerate(train_loader,0):
    inputs,target = data
    show_images(inputs,target,target)
    break
if __name__ =='__main__':
    for epoch in range(10):
        for batch_idx , data in enumerate(train_loader,0):
            inputs,target = data
            # print (inputs.shape)
            
```

```python
class classificationModel(torch.nn.Module):
    def __init__(self):
        super(classificationModel, self).__init__()
        self.linear1 = torch.nn.Linear(784, 512)
        self.linear2= torch.nn.Linear(512, 256)
        self.linear3= torch.nn.Linear(256, 128)
        self.linear4= torch.nn.Linear(128, 64)
        self.linear5= torch.nn.Linear(64, 10)

    def forward(self, x):
        x = x.view(-1,784)
        x = F.relu(self.linear1(x))
        x = F.relu(self.linear2(x))
        x = F.relu(self.linear3(x))
        x = F.relu(self.linear4(x))
        x = self.linear5(x)        
        return x

model = classificationModel()

criterion = torch.nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=1e-3,momentum=0.5)
```

```python
def train(rpoch):
    running_loss= 0.0
    for batch_idx,data in enumerate(train_loader,0):
        inputs,target = data
        optimizer.zero_grad() #优化器清零
        outputs = model(inputs)
        loss = criterion(outputs, target)
        loss.backward() 
        optimizer.step()
        running_loss += loss.item()
        if batch_idx %300 ==299:
            print('[%d,%5d] loss : %.3f' % (epoch, batch_idx+1, running_loss/300))
            running_loss = 0 

def test():
    correct =0
    total = 0
    with torch.no_grad():
        for data in test_loader:
            images,labels = data
            outputs = model(images)
            _,predicted = torch.max(outputs.data,dim = 1 ) #dim=0 :row
            total += labels.size(0)
            correct+= (predicted==labels).sum().item()
    print('acc = ',(100 * correct)/total)
```

```python
if __name__ == '__main__':
    for epoch in range(100000):
        train(epoch)
        test()
```

```python
with torch.no_grad():
    for data in test_loader:
        images,labels = data
        outputs = model(images)
        _,predicted = torch.max(outputs.data,dim = 1 ) #dim=0 :row
        # print(predicted)
        show_images(images,labels,predicted)
        break
```

```python
import torchvision
batch_size = 64
transform  = transforms.Compose([
    transforms.ToTensor(), #convert pil image to tensor
    transforms.Normalize((0.1307,),( 0.3081,)) # mean & std,数字是对mnist计算出的经验值
])
train_dataset = datasets.MNIST(root='../dataset/mnist',train = True,transform=transform,download = True)
test_dataset = datasets.MNIST(root='../dataset/mnist',train = False,transform=transform,download = True)
train_loader = DataLoader(dataset= train_dataset,batch_size= batch_size ,shuffle= True)
test_loader = DataLoader(dataset= test_dataset,batch_size= batch_size ,shuffle= True)
```

```python
class Net(torch.nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv_layer1 = torch.nn.Conv2d(1,10,kernel_size=5,padding=0,stride=1,bias = False)
        self.conv_layer2 = torch.nn.Conv2d(10,20,kernel_size=5,padding=0,stride=1,bias = False)
        self.maxPooling_layer = torch.nn.MaxPool2d(kernel_size= 2)
        self.fc = torch.nn.Linear(320,10)

    def forward(self, x):
        batch_size = x.size(0)
        x = self.conv_layer1 (x)
        # print(x.shape)
        x = F.relu(self.maxPooling_layer(x))
        # print(x.shape)
        x = self.conv_layer2 (x)
        # print(x.shape)
        x = F.relu(self.maxPooling_layer(x))
        # print(x.shape)
        x = x.view(batch_size,-1) #flatten
        # print(x.shape)
        x = self.fc(x) #做交叉熵损失，最后一层不做激活       
        return x

model = Net()

criterion = torch.nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=1e-3,momentum=0.5)
```

```python
class NetMoreLayer(torch.nn.Module):
    def __init__(self):
        super(NetMoreLayer, self).__init__()
        self.conv_layer1 = torch.nn.Conv2d(1,128,kernel_size=5,padding=2,stride=1,bias = True)
        self.conv_layer2 = torch.nn.Conv2d(128,64,kernel_size=5,padding=2,stride=1,bias = True)
        self.conv_layer3= torch.nn.Conv2d(64,32,kernel_size=3,padding=2,stride=1,bias = True)
        self.maxPooling_layer = torch.nn.MaxPool2d(kernel_size= 2)
        self.fc1 = torch.nn.Linear(512,128)
        self.fc2 = torch.nn.Linear(128,64)
        self.fc3 = torch.nn.Linear(64,10)

    def forward(self, x):
        batch_size = x.size(0)
        x = self.conv_layer1 (x)
        # print(x.shape)
        x = F.relu(self.maxPooling_layer(x))
        # print(x.shape)
        x = self.conv_layer2 (x)
        # print(x.shape)
        x = F.relu(self.maxPooling_layer(x))
        # print(x.shape)
        x = F.relu(self.maxPooling_layer(self.conv_layer3(x)))        
        # print(x.shape)
        x = x.view(batch_size,-1) #flatten
        # print(x.shape)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x) #做交叉熵损失，最后一层不做激活       
        return x

model = NetMoreLayer()

criterion = torch.nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=1e-3,momentum=0.5)
```

```python
def train(rpoch):
    running_loss= 0.0
    for batch_idx,data in enumerate(train_loader,0):
        inputs,target = data
        
        optimizer.zero_grad() #优化器清零
        outputs = model(inputs)
        loss = criterion(outputs, target)
        loss.backward() 
        optimizer.step()
        running_loss += loss.item()
        if batch_idx %300 ==0:
            print('[%d,%5d] loss : %.3f' % (epoch, batch_idx+1, running_loss/300))
            running_loss = 0 

def test():
    correct =0
    total = 0
    with torch.no_grad():
        for data in test_loader:
            images,labels = data            
            outputs = model(images)
            _,predicted = torch.max(outputs.data,dim = 1 ) #dim=0 :row
            total += labels.size(0)
            correct+= (predicted==labels).sum().item()
    print('acc = ',(100 * correct)/total)
```

```python
if __name__ == '__main__':
    for epoch in range(1000):
        train(epoch)
        test()
        if epoch % 10 ==0:
            with torch.no_grad():
                for data in test_loader:
                    images,labels = data                    
                    outputs = model(images)
                    _,predicted = torch.max(outputs.data,dim = 1 ) #dim=0 :row
                    # print(predicted)
                    show_images(images,labels,predicted)
                    break
            
       
```
